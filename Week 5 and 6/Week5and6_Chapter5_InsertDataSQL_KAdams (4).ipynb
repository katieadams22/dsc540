```python
# Created a query with the above 6 fields (Name, Address, City, State, Zip, Phone Number) with appropriate field types by
# using in-memory SQLite database using Python's built-in sqlite3 driver

import sqlite3

query = """
CREATE TABLE test
(Name VARCHAR(20), 
Address VARCHAR(20), 
City VARCHAR(20), 
State VARCHAR(20), 
Zip int, 
Phone Number VARCHAR(20)
);"""
con = sqlite3.connect(':memory:')
con.execute(query)
con.commit()
```
