## Convert between string and date time

```python
from datetime import datetime
met = pd.read_csv("MetObjects.csv") # new variable met that reads the other dataset available in this assignment
met['Object Begin Date'] = met['Object Begin Date'].astype(str) # change the AccessionYear column from float to string

datetime_str = met['Object Begin Date'] # variable that contains a date string that is not a date object
datetime_str.head(5)
```

```python
# used the to_datetime method to parse many different kinds of date representations
datetime_obj = pd.to_datetime(datetime_str, errors = 'coerce') # used errors = 'coerce' to force the dates which are out of bounds
datetime_obj.head(5)
```

## Generate date range

```python
# pandas.date_range is responsible for generating  a DatetimeIndex with an indicated length according to a particular frequency
index = pd.date_range(met.at[2, "Object Begin Date"], periods = 20) # created index variable using 1 cell in dataset and passed a number of periods to generate
index
```

## Frequencies and date offsets

```python
from pandas.tseries.offsets import Hour, Minute #import hour and minutes from pandas
hour = Hour() # hourly frequency represented with the Hour class
hour
```
